Create the data visualization using the matplotlib.
 Visualize the Gender of Passengers using the Bar graph.
 Visualize the Survival Count of Passengers using the Bar graph.
 Visualize the Age of Passengers using the Bar/Histogram graph.
 Visualize the comparison of Age and Fare of Passengers using the Scatterplot.


```python
import matplotlib.pyplot as plt
df['Sex'].value_counts().plot(kind='bar',ylabel="Number of People",title="Visualization on Gender")
```

```python
df['Survived'].value_counts().plot(title="Visualization on Survival",kind='bar',ylabel="Number of People",color=["blue","green"]).legend(['Deceased','Survived'])
```

```python

plt.hist(df['Age'].value_counts())
plt.title("Age of passengers")
```

```python
df.plot.scatter(x='Age', y='Fare',title="Comparison Between Age and Fare")
```
