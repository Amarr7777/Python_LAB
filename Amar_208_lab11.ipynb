Perform the Exploratory Data Analysis on your domain-based dataset and demonstrate
the retrieved insights using “Matplotlib” modules. Visualize hidden insights using appropriate
plots (graphs) [Usage of line plot and scatter plot are mandatory]

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the CSV file
data = pd.read_csv('../juptyer-notebook/csv/salaries_final.csv')

# Create a dataframe from the data
df = pd.DataFrame(data)
df = df.head(50)
print(df)
# Create a line graph
plt.plot(df['Primary Job Title'], df['Base Pay'])
plt.xlabel('Position Title')
plt.ylabel('Base Pay')
plt.title('Line Graph of Position Title vs. Given Name')
plt.show()

# Create a bar graph
plt.bar(df['Primary Job Title'], df['Base Pay'])
plt.xlabel('Position Title')
plt.ylabel('Salary Paid')
plt.title('Bar Graph of Position Title vs. Salary Paid')
plt.show()

# Create a scatter plot
plt.scatter(df['Primary Job Title'], df['Base Pay'])
plt.xlabel('Position Title')
plt.ylabel('Salary Paid')
plt.title('Scatter Plot of Position Title vs. Salary Paid')
plt.show()

# Create a correlation graph or heat map
plt.imshow(df.corr(), cmap='hot')
plt.colorbar()
plt.title('Correlation Graph of Faculty Data')
plt.show()
```
